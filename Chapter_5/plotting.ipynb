```python
from scipy.optimize import minimize
from scipy.optimize import basinhopping
import matplotlib.pyplot as plt
import numpy as np
import os
import elementaries as el
```

```python
# ,'ordfullQSA':2
methods=dict([('locQSA',0)])

methods.update([('fullQSA',0)])
methods.update([('ordfullQSA',0)])
methods.update([('2locQSA',0)])
methods.update([('revQSA',0)])
method='SLSQP'
J=6
N=8
directory="export/"+method+"/N="+str(N)+"/J="+str(J)+"/"
```

```python
values=dict()

for element in methods:
    values.update([(element,np.loadtxt(fname=directory+element+"_values_J="+str(J)+"_N="+str(N)+".txt") )])
```

```python
total_measurements=dict()

for element in methods:
    total_measurements.update([(element,np.loadtxt(fname=directory+element+"_total_measurements_J="+str(J)+"_N="+str(N)+".txt") )])
```

```python
plt.rc('text', usetex=True)

# fig, ax = plt.subplots()

# ax.set_yscale('log')
mingates=0
maxgates=30
scale=3
s=scale*120
plt.xlabel('The number of Hamiltonian measurements', fontsize=scale*40)
plt.ylabel('$|E-E_{GS}|/|E_{GS}|$', fontsize=scale*40)
for element in values:
    plt.plot(list(range(mingates,maxgates)),total_measurements[element][mingates:maxgates])
    plt.scatter(list(range(mingates,maxgates)),total_measurements[element][mingates:maxgates],label=element,s=s)
# plt.text(1,0.01, '$H$ of TFIM type, $N=8, J/h=0.15$', {'color': 'black', 'fontsize': scale*40}, verticalalignment='bottom')
plt.legend(bbox_to_anchor=(1,1), loc=2, borderaxespad=0.,fontsize=scale*30)
plt.tick_params(direction='in')
# ax.set_xticklabels(1, rotation=0, fontsize=8)

plt.rcParams['xtick.labelsize']=scale*30
plt.rcParams['ytick.labelsize']=scale*30

fig=plt.gcf()
axes=plt.gca()
# axes.set_yscale('log')
fig.set_size_inches(scale*10, scale*5)
print(fig)
fig.savefig('N=8_J=0.15.pdf', bbox_inches='tight')
fig.show()
```
